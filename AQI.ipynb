# Exploratory Data Analysis

```python
#Plot of AQI value
plt.figure(figsize=(10, 6))
plt.hist(data['AQI Value'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('AOI Value')
plt.ylabel('Frequency')
plt.title('Histogram of AQI Value')
plt.grid(True)
plt.show()
```

Most of the cities falls under 100 AQI values (Good and Moderate), but some of are with more than 400, Which is hazardous for health concern.

```python
#CO AQI Value
plt.figure(figsize=(10, 6))
plt.hist(data['CO AQI Value'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('CO AQI Value')
plt.ylabel('Frequency')
plt.title('Histogram of CO AQI Value')
plt.grid(True)
plt.show()
```

Most of the cities comes under less than 20 CO value (0-50ppm) is indicate good. 

```python
#Ozone AQI Value
plt.figure(figsize=(10, 6))
plt.hist(data['Ozone AQI Value'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Ozone AQI Value')
plt.ylabel('Frequency')
plt.title('Histogram of Ozone AQI Value')
plt.grid(True)
plt.show()
```

Ozone in stratosphare is good for Earth, but on ground it is not. 0-100 value in atmosphare is acceptable, beyond 100 level of ozone is very unhealthy for respiratory diseases (chest pain, coughing, throat irritation, and congestion).

```python
#NO2 AQI Value
plt.figure(figsize=(10, 6))
plt.hist(data['NO2 AQI Value'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('NO2 AQI Value')
plt.ylabel('Frequenccy')
plt.title('Histogram of NO2 AQI Value')
plt.grid(True)
plt.show()
```

NO2 gas beyond 20 value is highly impact on environment, it is very harmfull for lung disease. 

```python
#PM2.5 AQI Value
plt.figure(figsize=(10,6))
plt.hist(data['PM2.5 AQI Value'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

PM2.5 concentration is directly proportional to AQ in atmosphare, it is cause heart disease, asthma and low birth weight.

```python
data.groupby('Country')['AQI Value'].count().sort_values(ascending=False).head(15).sort_values(ascending=True).plot(kind='barh')
plt.title('Most represented countries\nwithin the dataset', loc='left', fontsize=25)
plt.xlabel('Number of entries')
plt.grid(True)
plt.show()
```

Most number of cities in table are from USA, Germany and Brazil.

```python
#Top 30 Polutted Cities by AQI

top_30 = data.nlargest(30, 'AQI Value').sort_values(by='AQI Value', ascending=True)
plt.figure(figsize=(10, 6))
plt.barh(top_30['City'], top_30['AQI Value'], color='skyblue', edgecolor='black')
plt.title('Bar Graph of Top 30 Maximum Values of AQI')
plt.xlabel('AQI Value')
plt.ylabel('City')
plt.grid(True)
plt.show()
```

```python
#Top 30 cleanest Cities
bottom_30 = data.nsmallest(30, 'AQI Value').sort_values(by='AQI Value', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(bottom_30['City'], bottom_30['AQI Value'], color='skyblue', edgecolor='black')
plt.title('Bar Graph of top 30 cleanest Cities by AQI')
plt.xlabel('AQI Value')
plt.ylabel('City')
plt.grid(True)
plt.show()
```

MAcas (Ecuador), Tari (Papua New Guinea) and Azogues (Ecuador) cities are ,cleanest cities in the world.

```python
data.groupby('Country')['AQI Value'].mean().nlargest(20)
```

```python
#Top 20 Polutted Country
plt.figure(figsize=(10, 6))
mean_AQI=data.groupby('Country')['AQI Value'].mean().nlargest(20)
mean_AQI.plot(kind='line',style='*')
plt.title('Top 20 Countries by AQI value (Mean of Cities AQI)')
plt.xlabel('Countries')
plt.ylabel('Mean_AQI')
plt.grid(True)
plt.show()
```

```python
#Top 20 Cleanest Country
data.groupby('Country')['AQI Value'].mean().nsmallest(20)
```

```python
#Top 20 cleanest Country on graph 
plt.figure(figsize=(10, 6))
mean_AQI=data.groupby('Country')['AQI Value'].mean().nsmallest(20)
mean_AQI.plot(kind='line',style='*')
plt.title('Bottom 20 Countries by AQI value (Mean of Cities AQI)')
plt.xlabel('Countries')
plt.ylabel('Mean_AQI')
plt.grid(True)
plt.show()
```
