# Exploratory Data Analysis

```python
#CO AQI Value 
fig=sns.scatterplot(x='CO AQI Value', y='AQI Value', data=data)
fig
```

CO value increase a little this will impact the AQI drastically, which is cause of headache, dizziness and loss of consciousness. 

```python
#Ozone AQI Value
fig2=sns.scatterplot(x='Ozone AQI Value', y='AQI Value', data=data)
fig2
```

As Ozone concentration increase AQI value will increase, respiratory disease may increase among the people.

```python
#NO2 AQI Value
fig3=sns.scatterplot(x='NO2 AQI Value', y='AQI Value', data=data)
fig3
```

NO2 gas highly impact on environment, it is very harmfull for lung disease. 

```python
#PM2.5 AQI Value
fig4=sns.scatterplot(x='PM2.5 AQI Value', y='AQI Value', data=data)
fig4
```

PM2.5 concentration is directly proportional to AQI value, it is cause heart disease, asthma and low birth weight.

```python
#Top 20 Polutted Cities
data.nlargest(30, ['AQI Value'])
```

```python
#Top 30 cleanest Cities
data.nsmallest(30, ['AQI Value'])
```

```python
#Top 20 Polutted Country
data.groupby('Country')['AQI Value'].mean().nlargest(20)
```

```python
Avg_aqi=data.groupby('Country')['AQI Value'].mean().nlargest(20)
```

```python
#Top 20 Polutted Country on graph 
Avg_aqi.plot(kind='line',style='*')
plt.xlabel('Countries')
plt.ylabel('Avg_AQI')
plt.show()
```

```python
#Top 20 Cleanest Country
cl=data.groupby('Country')['AQI Value'].mean().nsmallest(20)
```
